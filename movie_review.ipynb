```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
```

```python
from nltk import pos_tag
s = "better"
pos_tag([s])
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
import string
punctuations = list(string.punctuation)
stop = stop + punctuations
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document),category)for document, category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[:1500]
testing_documents = documents[1500:]
```

```python
all_words = list()
for doc in documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
```

```python
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
len(features)
```

```python
def get_feature_dict(words):
    current_words = {}
    words_set = set(words)
    for w in features:
        current_words[w] = w in words_set
    return current_words
```

```python
get_feature_dict(training_documents[0][0])
```

```python
training_data = [(get_feature_dict(document),category) for document, category in training_documents] 
testing_data = [(get_feature_dict(document),category) for document, category in testing_documents]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
